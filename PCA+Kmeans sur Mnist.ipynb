```python
import numpy as np
from numpy import array
from numpy import mean
from numpy import cov
from numpy.linalg import eig
import pandas as pd
import matplotlib.pyplot as plt
import plotly.plotly as py
import plotly.graph_objs as go
from IPython.display import Image
%matplotlib inline
```

```python
from sklearn.datasets import fetch_mldata
mnist = fetch_mldata('MNIST original', data_home='datasets/')

# Convert sklearn 'datasets bunch' object to Pandas DataFrames
y = pd.Series(mnist.target).astype('int').astype('category')
X = pd.DataFrame(mnist.data)
```

```python
new_column_name = []
for num in range(X.shape[1]):
    tem = 'pixel' + str(num)
    new_column_name.append(tem)
X.columns = new_column_name
```

```python
X.shape
```

```python
X['label'] = y
```

```python
X = X.sample(70000)
```

```python
y = X["label"]  # dependent variable
X = X.drop(["label"], axis = 1) #independent variable
```

```python
X.head()
```

```python
#PCA valeurs propres et vecteurs propres 
mean_vec = np.mean(X, axis=0)
cov_mat = (X- mean_vec).T.dot((X - mean_vec)) / (X.shape[0])
eig_vals, eig_vecs = np.linalg.eig(cov_mat)

print('Eigenvectors \n%s' %eig_vecs)
print('\nEigenvalues \n%s' %eig_vals)
```

```python
eig_vals = np.real(eig_vals)
eig_vecs = np.real(eig_vecs)
print('Eigenvectors \n', eig_vecs)
print('\nEigenvalues \n', eig_vals)
```

```python
#PCA valeurs propres sorted et vecteurs propres sorted
eig_vals_indexed = []
for i in range(len(eig_vals)):
    eig_vals_indexed.append((i,abs(eig_vals[i])))
```

```python
def getValue(l):
    return l[1]

eig_vals_sorted = sorted(eig_vals_indexed, key=getValue,reverse= True)
```
