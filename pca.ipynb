```python
import numpy as np
from numpy import array
from numpy import mean
from numpy import cov
from numpy.linalg import eig
import pandas as pd
```

```python
df = pd.DataFrame()
df['V1'] = [1,2,3]
df['V2'] = [1,4,6]
df['v3'] = [1,6,7]
df['V4'] =[1,5,7]
df['V5'] = [9,5,6]
```

```python
df
```

```python
a = list(df.mean())
```

```python
for i in range(df.shape[1]) : 
    df.iloc[:,i] = df.iloc[:,i]-a[i]
```

```python
df
```

```python
mean_vec = np.mean(df, axis=0)
cov_mat = (df - mean_vec).T.dot((df - mean_vec)) / (df.shape[0]-1)
```

```python
cov_mat
```

```python
cov_mat = np.cov(df.T)

eig_vals, eig_vecs = np.linalg.eig(cov_mat)

print('Eigenvectors \n%s' %eig_vecs)
print('\nEigenvalues \n%s' %eig_vals)
```

```python
from sklearn.datasets import fetch_mldata
mnist = fetch_mldata('MNIST original', data_home='datasets/')

# Convert sklearn 'datasets bunch' object to Pandas DataFrames
y = pd.Series(mnist.target).astype('int').astype('category')
X = pd.DataFrame(mnist.data)
```

```python
def pca(df):
    mean_vec = np.mean(df, axis=0)
    cov_mat = (df - mean_vec).T.dot((df - mean_vec)) / (df.shape[0])
    cov_mat = np.cov(df.T)

    eig_vals, eig_vecs = np.linalg.eig(cov_mat)

    print('Eigenvectors \n%s' %eig_vecs)
    print('\nEigenvalues \n%s' %eig_vals)
```
